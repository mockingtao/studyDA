```python
# （1）分析出女性市委书记占比最高的TOP10的城市？并制作柱状图

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('/Users/matao/Study/data_analyst/python体验课学习/data_test2.csv', encoding='gbk', engine='python')
data = data[data['性别'].notnull()]
#print(data[120:130])

data_re = data[['地市级政区名称','党委书记姓名','性别']].drop_duplicates()
#print(data_re.head())

pt = pd.crosstab(data_re['地市级政区名称'],data_re['性别'])
pt['女性占比'] = pt['女'] / (pt['女']+ pt['男'])
pt2 = pt.sort_values(by=['女性占比'], ascending = False)
pt2['女性占比百分数'] = pt2['女性占比'] *100
print(pt2.head(10))
x = pt2.head(10).index.tolist()
y = pt2[:10]['女性占比百分数'].tolist()
fig1 = plt.figure(figsize=(10,4))

sns.barplot(x, y, palette="BuPu_r")
plt.title("Percentage of Women Leaders")
plt.ylabel('Per %')
sns.despine(bottom=True)
plt.show()
```

```python
# （2.1）分析出整体女性市委书记占比结构，并制作箱线图

sns.set(context="notebook")
x2 = ['Percentage of Women leaders_structure']
y2 = pt2['女性占比'].tolist()
print(y2)

fig2_1 = plt.figure(figsize=(6,4))
sns.boxplot(data=y2, palette='Greens')
plt.title('Percentage of women leaders')
plt.ylabel('Per %')

sns.despine(bottom=False)
plt.show()


```

```python
# （2.2）分析出整体女性市委书记占比结构，并制作小提琴图

x3 = ['Percentage of women leaders_structure']
y3 = pt2['女性占比百分数'].tolist()

fig2_2 = plt.figure(figsize=(6, 4))
sns.violinplot(data=y3, palette='Oranges')
plt.title('Percentage of women leaders')
plt.ylabel('Per %')

sns.despine(bottom=True)
plt.show()
```
