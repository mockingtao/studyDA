```python
'''问题一_性别构成'''


# 绘制图表1：不同省份女性市委书记占比
# Seaborn可视化 - 柱状图

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="white", context="notebook")
fig = plt.figure(figsize=(10,4))

data = pd.read_csv("/Users/matao/Study/data_analyst/python体验课学习/data_test2.csv", encoding='gbk', engine='python')
data_gender2 = data[['省级政区名称','性别']]
data_gender2_re = data_gender2[data_gender2['性别'].notnull()]
pt = pd.crosstab(data_gender2_re['省级政区名称'], data_gender2_re['性别'])
pt['女性占比'] =  pt['女'] / (pt['女'] + pt['男'])
pt2 = pt.sort_values(by = ['女性占比'], ascending = False)
print('\n====第①个=================')
print(pt2.head())

print('\n====第②个=================')
x = ['Liaoning','Shanxi','Jilin','Shanxi_m','Sichuan','Anhui','Jiangxi','Shandong','Fujian','Yunnan']
pt2['女性占比百分数'] = pt2['女性占比'] *100
print(pt2.head())

print('\n====第③个=================')
y = pt2[:10]['女性占比百分数'].tolist()
sns.barplot(x, y, palette="BuPu_r")
plt.title("Percentage of Women Leaders")
plt.ylabel('Per %')

sns.despine(bottom=True)
plt.show()


```

```python
# 绘制图表2：女性市委书记占比结构
# Seaborn可视化 - 箱线图
sns.set(context="notebook")

x2 = ['Percentage of Women leaders_structure']
y2 = pt2['女性占比'].tolist()

fig_2 = plt.figure(figsize=(6, 4))
sns.boxplot(data=y2, palette='Greens')
plt.title('Percentage of Women leaders')
plt.ylabel('Per %')

sns.despine(bottom=True)
plt.show() 
```

```python
# 绘制图表2：女性市委书记占比结构
# Seaborn可视化 - 小提琴图

x3 = ['Percentage of Women leaders_structure']
y3 = pt2['女性占比百分数'].tolist()
#print(pt2)
print(y3)

fig = plt.figure(figsize=(6, 4))
sns.violinplot(data=y3, palette='Oranges')
plt.title('Percentage of Men leaders')
plt.ylabel('Per %')

sns.despine(bottom=True)
plt.show()
```

```python
'''
问题三_任期情况
'''
```

```python
# 绘制图表1：任期与出生年份关系
# Seaborn可视化 - 混合散点图

data_term = data[['年份','党委书记姓名','出生年份']]
data_term_re = data_term[data_term['出生年份'].notnull()]
year_max = data_term_re[['出生年份','年份']].groupby(data_term_re['党委书记姓名']).max()
year_max.rename(columns={'年份':'年份max'}, inplace = True)
year_max['姓名'] = year_max.index
#print(year_max.head())

year_min = data_term_re[['出生年份','年份']].groupby(data_term_re['党委书记姓名']).min()
year_min.rename(columns={'年份':'年份min'}, inplace = True)
year_min['姓名'] = year_min.index
#print(year_min.head())

data_term_fin = pd.merge(year_max,year_min)
#print(data_term_fin.head())
data_term_fin['任期'] = data_term_fin['年份max'] - data_term_fin['年份min']
#print(data_term_fin.head())


sns.set(style="ticks", context="notebook")
fig = plt.figure(figsize=(6, 5))
x = data_term_fin['任期']
y = data_term_fin['出生年份']

sns.jointplot(x, y, alpha=0.2)
sns.jointplot(x, y, kind="hex")

plt.show()


```

```python
# 绘制图表2：任期与出生年份关系 - 热图
# Seaborn可视化 - 热图

sns.set(style="ticks", context="notebook")
fig = plt.figure(figsize = (10,6))
df = pd.crosstab(data_term_fin['任期'], data_term_fin['出生年份'])
print(df)
sns.heatmap(df, annot=True, fmt="d", linewidths=0.2, linecolor='k')
plt.show()
```
