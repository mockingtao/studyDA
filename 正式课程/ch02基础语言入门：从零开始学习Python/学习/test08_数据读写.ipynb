```python
'''
【课程8.3】  文件的读取与写入

os 模块：提供了非常丰富的方法用来处理文件和目录

'''
```

```python
# ANSI --- gbk
# UTF-8 --- 'utf8'

# 文件读取 - read

f = open('C:\\Users\\Hjx\\Desktop\\text.txt', 'r')
print(f.read())
# 直接读取（这里光标默认会在最末尾）

f.seek(0)
print(f.read(2))
# f.read(n):n代表读取多少个字符

f.seek(0) #每次读取之前，可以调这个函数。很有用！！！！！！！！！！！！
print(f.readline())
print(f.readline())
print(f.readline(4))
# 读取行到字符串，一次性读取一行
# f.readline(n)：读取该行的前n个字符

f.seek(0)
for line in f.readlines():
    print(type(line),line)
# 遍历一个文件：for语句+f.readlines()

```

```python
# 小作业
# 利用百度POI小插件爬取一些poi数据，然后存成txt，再用python读取，编写成一个json形式（列表字典）：
# [{'name':'...', 'lng':..., 'lat':..., 'address':'...'},{...},...,]


test_file = open('data_hz_university.txt','r')
m = []
test_file.seek(0)
n=0
for line in test_file.readlines():
    n+=1
    print(line)

    str1 = line.split(':')
    name = str1[0]
    information = str1[1]
    str2 = information.split(',')
    lng = str2[0]
    lat = str2[1]
    adderss = str2[2]
    data = [['name',name],['lng',lng],['lat',lat],['adderss',adderss]]
    m.append(dict(data))
print(m)
```

```python
# 文件写入 - write
path='data_testWrite2.txt'
f = open(path, 'w', encoding = 'utf8')
f.write('hello world!') # write() argument must be str, not list. 可见，write只能穿字符串，所以肯定是只有一个元素的。
#  write() takes exactly one argument (2 given)  --- 也确实，只能写一个元素，所以，这个时候就用到了 writelines()
# 执行后直接创建“test_write.txt”文件，并写入“hello world！”
# 12代表输入了12个字符，但是文件内真的有写入吗？打开看看

f.close()
# 为什么没有？因为刚才是在内存里运行，我们需要再加上close才能写入

path='data_testWrite3.txt'
f = open(path, 'w', encoding = 'utf8')
lst = ['a','b','c','d','e']
f.writelines(lst)
f.close()
# f.writelines(list)：依次把列表元素写入，但这里有没有提行？
# 如果想每一个元素提行怎么办？ → 每个元素后都有"\n"，可以使用“遍历”，如下
```

```python
path='data_testWrite4.txt'
f = open(path, 'w', encoding = 'utf8')
lst = ['a','b','c','d','e']

for i in range(len(lst)):
    lst[i] = lst[i] + '\n'

f.writelines(lst)
f.close()
```

```python
'''
【课程8.4】  pickle模块的运用---- 优点是：不会改变数据结构，如果存储为text，则会改变数据机构

pickle提供了一个简单的持久化功能。可以将对象以文件的形式存放在磁盘上。

python的pickle模块实现了基本的数据序列和反序列化

通过pickle模块的序列化操作我们能够将程序中运行的对象信息保存到文件中去，永久存储

通过pickle模块的反序列化操作，我们能够从文件中创建上一次程序保存的对象。

pickle.dump() / pickle.load()

'''
```

```python
# 存储：pickle.dump(obj, file, [,protocol])

import pickle

data = {'a':[1,2,3,4], 'b':('string','abc'), 'c':'hello'}
print(data)
# 创建一个字典变量data

pic = open( 'C:\\Users\\Hjx\\Desktop\\data.pkl', 'wb')
# 以二进制来存储：rb, wb, wrb, ab

pickle.dump(data,pic)
pic.close()
# 将一个字典数据存成了pkl文件
```

```python
# 读取：pickle.load(file)

f = open( 'C:\\Users\\Hjx\\Desktop\\data.pkl', 'rb')

st = pickle.load(f)
print(st)
# 直接读取pkl文件内的数据，该数据为一个字典
```
