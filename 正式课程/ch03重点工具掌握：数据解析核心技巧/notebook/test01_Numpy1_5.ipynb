```python
'''
【课程1.5】  Numpy随机数

numpy.random包含多种概率分布的随机样本，是数据分析辅助的重点工具之一

'''
```

```python
# 随机数生成
import numpy as np
samples = np.random.normal(size=(4,4))
print(samples)
# 生成一个标准正太分布的4*4样本值
```

```python
# numpy.random.rand(d0, d1, ..., dn)：生成一个[0,1)之间的随机浮点数或N维浮点数组 —— 均匀分布

import matplotlib.pyplot as plt
% matplotlib inline

a = np.random.rand() #生成一个随机浮点数
print(a, type(a))

b = np.random.rand(4) # 生成一个形状为4的一维数组
print(b, type(b))

c = np.random.rand(2,3) # 生成一个形状为2*3的二维数组，注意这里不是((2,3))...这样说的目的，是因为之前创建的时候用的就是zeros((2,3))
print(c, type(c))

samples1 = np.random.rand(1000)
samples2 = np.random.rand(1000)
plt.scatter(samples1, samples2) # 生成1000个均匀分布的样本值，samples1和sample22分别为x，y值。  两个轴的数据对应一个点。
```

```python
#  numpy.random.randn(d0, d1, ..., dn)：生成一个浮点数或N维浮点数组 —— 正态分布

samples1 = np.random.randn(1000)
samples2 = np.random.randn(1000)
plt.scatter(samples1, samples2)
```

```python
# numpy.random.randint(low, high=None, size=None, dtype='l')：生成一个整数或N维整数数组
# 若high不为None时，取[low,high)之间随机整数，否则取值[0,low)之间随机整数，且high必须大于low 
# dtype参数：只能是int类型  

a = np.random.randint(2) # low=2,生成一个[0,1]之间的随机整数
print(a)

b = np.random.randint(2, size=5) # low=2, size=5 :生成一个有5个元素，每个元素0~1之间的一维数组
print(b)

c = np.random.randint(2,6,size=5) # low=2, high=6,size=5: 生成一个有5个元素，每个元素2~5之间的一维数组
print(c)

d = np.random.randint(2, size=(2,3)) # low=2,size=(2,3) : 生成一个2*3 的整数数组，元素范围0~1随机整数
print(d)

e = np.random.randint(2,6,(2,3)) # low=2, high=6, size=(2,3):生成一个2*3的整数数组，元素范围2~5随机整数
print(e)
```

```python
# 作业1：请按照要求创建数组ar，再将ar[:2,:2]的值改为[0,1)的随机数

hw_ar1 = np.arange(25,dtype=float).reshape(5,5)
print(hw_ar1)

hw_ar1_22 = hw_ar1[:2,:2]
print(hw_ar1_22)

hw_ar1[:2,:2] = np.random.rand(2,2)
print(hw_ar1_22)
print('---------------')
print(hw_ar1)
```
